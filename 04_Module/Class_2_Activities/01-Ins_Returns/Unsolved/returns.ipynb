# Calculating Stock Returns using Pandas DataFrames

## 1. Import Libaries and Dependencies

```python
# Import necessary libraries and dependencies
import pandas as pd
from pathlib import Path
%matplotlib inline
```

## 2. Create a Path to the File Using Pathlib

```python
# Use the Pathlib libary to set the path to the CSV
csvpath = Path('../Resources/MSFT.csv')
```

## 3. Read CSV in as DataFrame

```python
# Read in the CSV as a DataFrame
msft_df = pd.read_csv(
    csvpath,
    index_col='date',
    infer_datetime_format=True,
    parse_dates=True
)
```

## 4. Drop Extra Columns

```python
# Drop unnecessary columns
msft_df.drop(['volume', 'open', 'high', 'low'], axis=1, inplace=True)
```

## 5. Plot Daily Close

```python
# Plot the daily closing stock price of the S&P 500 over the last 5 years
msft_df.loc['2015-01-01' : '2020-01-01'].plot(legend=True, figsize=(10, 5), title='Closing Price of MSFT Over 5 Years', color='blue', label='MSFT', xlabel='date', ylabel='price')
```

## 6. Calculate Daily Returns

```python
# Calculate the daily return using the 'pct_change()' function
daily_returns = msft_df.pct_change()
```

## 7. Plot Daily Returns

```python
# Plot the daily returns of MSFT over the last 10 years
daily_returns.plot()
```

## 8. Calculate Cumulative Returns

```python
# Calculate the cumulative returns using the 'cumprod()' function
cumulative_returns_msft = (1 + daily_returns).cumprod()
cumulative_returns_msft
```

## 9. Plot Cumulative Returns

```python
# Plot the daily returns of the S&P 500 over the last 5 years
cumulative_returns_msft.loc['2015-01-01' : '2020-01-01'].plot()
```

# Alpaca Demo

```python
# Initial imports
import os
import requests
import pandas as pd
import seaborn as sys
from dotenv import load_dotenv
import alpaca_trade_api as tradeapi

%matplotlib inline
```

```python
# Load .env environment variables
load_dotenv()
```

```python
# Set Alpaca API key and secret
alpaca_api_key = os.getenv('ALPACA_API_KEY_ENV')
alpaca_secret_key = os.getenv('ALPACA_SECRET_KEY_ENV')
```

```python
# Verify that Alpaca key and secret were correctly loaded
print(type(alpaca_api_key))
print(type(alpaca_secret_key))
```

```python
# Create the Alpaca API object
alpaca = tradeapi.REST(
    alpaca_api_key,
    alpaca_secret_key,
    api_version="v2")
```

```python
# Format current date as ISO format
today = pd.Timestamp("2022-01-28", tz="America/New_York").isoformat()
```

```python
# Set the tickers
tickers = ['FB', 'TWTR']
```

```python
# Set timeframe to one day ('1D') for the Alpaca API
timeframe = "1D"
```

```python
# Get current closing prices for FB and TWTR
df_portfolio_today = alpaca.get_barset(
    tickers,
    timeframe,
    start = today,
    end = today
).df

# Display sample data
df_portfolio_today
```

```python
# Format start and end dates as ISO format for one year period
start_date = pd.Timestamp("2021-01-28", tz="America/New_York").isoformat()
end_date = pd.Timestamp("2022-01-28", tz="America/New_York").isoformat()
```

```python
# Set limit_rows to 1000
limit_rows = 1000
```

```python
# Get closing prices for FB and TWTR from the last year
df_portfolio_year = alpaca.get_barset(
    tickers,
    timeframe,
    start = start_date,
    end = end_date,
    limit = limit_rows
).df

# Display sample data
df_portfolio_year
```
